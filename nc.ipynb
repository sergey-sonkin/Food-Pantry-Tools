```python
import pandas as pd;
cols=['ID','Description','Package','Storage','Ext. Wt.','Ext. $$','Quantity','Date']
df1 = pd.DataFrame(columns=cols)
row={}
i=0
file_obj = open('C:\\NCC\\order.txt', 'r')
file_name=file_obj.readline().strip()
for line in file_obj:
    if ((i+1) % 7 != 0):
        row.update( {cols[i] : line.strip()} )  
        i=i+1 
 #       print(line.strip(),  end ="|"),
    else:
        row.update( {cols[i] : line.strip()} )  
        row.update( {cols[i+1] : file_name} )  
        df1=df1.append(row, ignore_index=True)
        i=0
 #       print(line.strip()),     
        
excel_file="C:\\NCC\\"+file_name+".xlsx"
csv_file="C:\\NCC\\"+file_name+".csv"
df1.to_excel(excel_file)
df1.to_csv(csv_file)
print(excel_file)
df1
```
